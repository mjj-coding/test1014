### Solution 2-1

```python
ListMonth=[10909,10910,10911,10912,11001,11002,11003,11004,11005,11006,11007,11008]

allYearData=pd.DataFrame()
    
for i in ListMonth:
    test=pd.read_csv('./data2/opendata{}M013.csv'.format(i))  #合併DataFrame
    allYearData=pd.concat([allYearData, test],axis=0)

allYearData.columns=population.columns                        #欄位名稱更換為中文
allYearData.drop(allYearData.index[0],axis=0,inplace=True)

allYearData['出生數']=allYearData['出生數'].astype(int)        #改整數型態
allYearData.iloc[:, 4:] = allYearData.iloc[:, 4:].astype(int)


print(allYearData.dtypes)
allYearData.head()
```

### Solution 2-2

```python
allYearDataGP=allYearData.groupby('區域別')
allYearDataGP.size()
```

### Solution 2-3

```python
for GPname in  allYearDataGP.indices.keys():
    birth=allYearDataGP.get_group(GPname).sum()['出生數']
    death=allYearDataGP.get_group(GPname).sum()['死亡數']
    birth_death=birth+death
    print('name: ', GPname)
    print('出生死亡加總: ', birth_death)
```
